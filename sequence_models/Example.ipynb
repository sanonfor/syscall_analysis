# SIMIAN Example Notebook

We always start by enabling matplotlib graphics and importing the `simian_lstm` prelude.

```python
%matplotlib inline
from simian_lstm.prelude import *
```

We train a selection of models with different layer configurations using hyperparameters sourced from [Kim et al.](https://arxiv.org/abs/1611.01726).

```python
train(
    dataset=adfa,
    tag="kim-replication",
    experiment={
        "embedding": -1,
        "layers": [[200], [400], [400, 400]],
        "learning_rate": 0.0001,
        "dropout": 0.5,
        "clipnorm": 10.0,
        "epochs": 30,
        "patience": 5,
    },
    model_version=2
)
```

Now we evaluate the models we trained in the last step, as well as their ensemble performance.

```python
evaluate("kim-replication", ensemble=True)
```
